```python
# SMIT - Artificial Intelligence & Data Science Course
# Project: Banking System Assignment (Functional Programming)
# Student Name: Jamal Nasir
# RollNumber: 335992

#Account List
accounts = {}

# Create an account fuction: asking name and initial balance. 

def create_account(name, initial_balance):
    if name in accounts:
        return f"Account for {name} already exists."
    if initial_balance < 0:
        return "Initial balance cannot be negative."
    accounts[name] = {
        'balance': initial_balance,
        'transactions': [f"Account created with initial balance: {initial_balance}-/Rs"]
    }
    return f"Account created successfully for {name}."


#Deposit money function: asking name and amount to deposit

def deposit(name, amount):
    if name not in accounts:
        return f"Account for {name} does not exist."
    if amount <= 0:
        return "Deposit amount must be positive."
    accounts[name]['balance'] += amount
    accounts[name]['transactions'].append(f"Deposited: {amount}-/Rs")
    return f"Deposited {amount}-/Rs to {name}'s account."

#Withdraw money function: asking name and amount to deposit
def withdraw(name, amount):
    if name not in accounts:
        return f"Account for {name} does not exist."
    if amount <= 0:
        return "Withdrawal amount must be positive."
    if amount > accounts[name]['balance']:
        return "Insufficient balance."
    accounts[name]['balance'] -= amount
    accounts[name]['transactions'].append(f"Withdrew: {amount}-/Rs")
    return f"Withdrew {amount}-/Rs from {name}'s account."

#Check balance function: asking name to view
def check_balance(name):
    if name not in accounts:
        return f"Account for {name} does not exist."
    return f"Current balance for {name}: {accounts[name]['balance']}-/Rs"


#Print a transaction statement functiona asking name to show statment
def print_statement(name):
    if name not in accounts:
        return f"Account for {name} does not exist."
    statement = f"Transaction statement for {name}:\n"
    statement += "\n".join(accounts[name]['transactions'])
    return statement


# Calling functions and printing results

print(create_account("Jamal", 1000))
print("")
print(deposit("Jamal", 500))
print("")
print(withdraw("Jamal", 200))
print("")
print(check_balance("Jamal"))
print("")
print(print_statement("Jamal"))


# Thank you
# Regards
# Jamal Nasir
# Rollnumber: 335992
```
